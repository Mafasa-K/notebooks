```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier,plot_tree
plt.style.use('ggplot')
```

```python
iris = load_iris()
```

```python
clf = DecisionTreeClassifier(max_depth=3)
```

```python
clf.fit(iris.data, iris.target)
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=list(iris.target_names), ax=ax)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Train a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Convert array to list for class_names
class_names = list(iris.target_names)

# Plot the decision tree
fig, ax = plt.subplots(figsize=(12, 12))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=class_names, ax=ax)
plt.show()
```
