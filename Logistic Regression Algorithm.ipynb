```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import warnings, random
plt.style.use('ggplot')
warnings.filterwarnings('ignore')
```

```python
X = np.random.normal(size = 50)
y = (X > 0).astype(float)
X[X > 0] *= 2
X += .3 * np.random.normal(size=50)
X= X.reshape(-1, 1)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X,y, color = 'r')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
lm_log = linear_model.LogisticRegression()
```

```python
lm_log.fit(X,y)
```

```python
X_ordered = np.sort(X, axis=0)
plt.scatter(X.ravel(), y, color='orange', zorder=20 , alpha = 0.5)
plt.plot(X_ordered, lm_log.predict_proba(X_ordered)[:,1], color='blue' ,linewidth = 3)
plt.show()
```
